# Euler

```python
METHOD = 'euler'
```

### Circular/120 Degree

```python
M = np.array([1,1,1])
G = 1
pos_init = np.zeros(shape=(3,3))
for i in range(3):
    phi = i*2*np.pi/3
    pos_init[i][0] = np.cos(phi)
    pos_init[i][1] = np.sin(phi)

acc_init = np.zeros(shape=(3,3))
for i in range(3):
    for j in range(i+1,3):
        r_diff = pos_init[j]-pos_init[i]
        r_norm = np.linalg.norm(r_diff)
        acc_init[i] += G*M[j]*r_diff/r_norm**3
        acc_init[j] -= G*M[i]*r_diff/r_norm**3

vel_init = np.zeros(shape=(3,3))
vel_abs = np.sqrt(-acc_init[0][0])
for i in range(3):
    phi = i*2*np.pi/3
    vel_init[i][0] = -vel_abs*np.sin(phi)
    vel_init[i][1] = vel_abs*np.cos(phi)
```

```python
euler_120_1 = NBodySim(N=3,DIM=3,POS=pos_init,VEL=vel_init,M=M,G=G)

T,dt = 10,0.01
euler_120_1.run(T,dt,method=METHOD)
euler_120_1.com_convert()

SAVEPATH = os.path.join(path,f'120deg_{dt}_{euler_120_1.method}.png')
euler_120_1.plot(LIM=None,savepath=SAVEPATH)
```

```python
euler_120_2 = NBodySim(N=3,DIM=3,POS=pos_init,VEL=vel_init,M=M,G=G)

T,dt = 10,0.001
euler_120_2.run(T,dt,method=METHOD)
euler_120_2.com_convert()

SAVEPATH = os.path.join(path,f'120deg_{dt}_{euler_120_2.method}.png')
euler_120_2.plot(LIM=None,savepath=SAVEPATH)
```

# Results

```python
sims = [euler_120_1,euler_120_2,euler_120_3,
        euler_8_1,euler_8_2,euler_8_3,
        lf_120_1,lf_120_2,lf_120_3,
        lf_8_1,lf_8_2,lf_8_3,
        hermite_120_1,hermite_120_2,hermite_120_3,
        hermite_8_1,hermite_8_2,hermite_8_3]

orbits = []
methods = []
dts = []
running_times = []
energy_diffs = []
i = 0
for sim in sims:
    orbits.append('120deg' if i < 3 else '8')
    methods.append(sim.method)
    dts.append(sim.dt)
    running_times.append(sim.running_time)
    energy_diffs.append(sim.energy_diff())
    i += 1
    if i == 6:
        i = 0

results = pd.DataFrame([orbits,methods,dts,running_times,energy_diffs],index=['Orbit','Method','Timestep','Running Time','Energy Difference']).T
results.sort_values(by=['Orbit','Method'])
```

# Animation

### Infinity Orbit

```python
M = np.array([1,1,1])
G = 1
pos_init = np.array([[0.9700436,-0.24308753,0],
                     [-0.9700436,0.24308753,0],
                     [0,0,0]])
vel_init = np.array([[0.466203685,0.43236573,0],
                     [0.466203685,0.43236573,0],
                     [-2*0.466203685,-2*0.43236573,0]])
anim_sim = NBodySim(N=3,DIM=3,POS=pos_init,VEL=vel_init,M=M,G=G)
T,dt = 10,0.01
anim_sim.run(T,dt,'hermite')
anim_sim.com_convert()
anim_sim.plot(LIM=1.5)
```

```python
from matplotlib.animation import FuncAnimation

pos = anim_sim.POS_COM
frames = pos.shape[0]

fig = plt.figure(figsize=(5,5))
ax = plt.axes(xlim=(-1.5,1.5),
              ylim=(-1.5,1.5))

trails = []
bodies = []
for i in range(anim_sim.N):
    trail, = ax.plot([],[],color=f'C{i}')
    body, = ax.plot([],[],color=f'C{i}',marker='o')
    trails.append(trail)
    bodies.append(body)

def update(frame):
    for i in range(anim_sim.N):
        trails[i].set_data(pos[max(0,frame-100):frame,i,0],pos[max(0,frame-100):frame,i,1])
        bodies[i].set_data([pos[frame,i,0]],[pos[frame,i,1]])
    return (*trails,*bodies)

anim = FuncAnimation(fig,update,frames=frames,interval=50,blit=True)
anim.save('8_dt001_hermite.mp4', fps=60)
plt.close()
```
